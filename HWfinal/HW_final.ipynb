Задача 44: В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего
из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без
get_dummies?

```python
import random
```

```python
import pandas as pd
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
ВАРИАНТ СДЕЛАТЬ КОПИЮ dataframe в варианте n-1
```

```python
new=data.copy(deep=True)
```

```python
new.head()
```

```python
new = new.rename (columns= {'whoAmI': 'robot'})
```

```python
new.head()
```

```python
new.replace(regex={r'robot': 1, 'human': 0})
```

ВАРИАНТ в текущем dataframe one hot в том виде как он был показал в google colab по ссылке

```python
data = data.rename (columns= {'whoAmI': 'robot'})
```

```python
data.replace(regex={r'robot': 1, 'human': 0}, inplace = True)
```

```python
data.loc[data['robot'] == 0, 'human'] = 1
```

```python
data.head()
```

```python
data.fillna(0)
```
